```python
import pandas as pd
import numpy as np
```

```python
#load review data
reviews = pd.read_csv('reviews.csv.gz', compression='gzip')
```

```python
reviews_clean = reviews.dropna()
```

```python
reviews_clean['comments'] = reviews_clean['comments'].str.lower()
```

```python
reviews_clean['comments'].head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')

reviews_clean['comments'] = reviews_clean['comments'].apply(lambda x:' '.join(x for x in x.split() if x not in stop))
```

```python
reviews_clean['comments'] = reviews_clean['comments'].str.replace(r'[^\w\s]+', '')
```

```python
#make sure punctuation is removed and everything is lowercase
reviews_clean['comments'][10]
```

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()

reviews_clean['comments'] = reviews_clean['comments'].apply(lambda x: ' '.join([st.stem(word) for word in x.split()]))
```

```python
freq = pd.Series(' '.join(reviews_clean['comments']).split()).value_counts()[:20]
freq
```
